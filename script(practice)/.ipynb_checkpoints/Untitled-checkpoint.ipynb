```python
import numpy as np#수학계산
import pandas as pd#데이터분석및 시각화
import matplotlib.pyplot as plt#시각화
import seaborn as sns
import os
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv1D, AveragePooling1D, MaxPooling1D
from tensorflow.keras.layers import Conv2D, AveragePooling2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv("C:/Users/jedy1/EnergyAI/temp_cloud_rain.csv",encoding = "utf-8",index_col=0)
df.head()
```

```python
temp = df["temp"].values
temp
```

```python
steps = 10#input 기간
forecast_terms = 15#예측 기간

x_train = np.array([temp[i:i+steps]for i in range(len(temp)-forecast_terms-steps)])
y_train = temp[steps:len(temp)-forecast_terms]

x_test = np.array([temp[-(steps+forecast_terms)+i:-forecast_terms+i] for i in range(forecast_terms)])
y_test = temp[-forecast_terms:]
```

```python
x_train.shape
```

```python
#column, channel
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1],1)
x_test = x_test.reshape(x_test.shape[0],x_test.shape[1],1)
```

```python
print("x_train:",x_train.shape)
print("y_train:",y_train.shape)
print("x_test:",x_test.shape)
print("y_test:",y_test.shape)
```

```python
model_1d = Sequential()
model_1d.add(Conv1D(filters=32,kernel_size=3,activation='relu',input_shape=(steps,1)))
model_1d.add(AveragePooling1D())

model_1d.add(Flatten())

model_1d.add(Dense(32,activation='relu'))
model_1d.add(Dense(4,activation='relu'))
model_1d.add(Dense(1))

adam = Adam()
model_1d.compile(loss = "mse", optimizer = adam)
```

```python
model_1d.summary()
#param = kernel_size*filters +bias of filters
```

```python
#학습
%time
model_1d_hist = model_1d.fit(x_train, y_train,epochs = 100,batch_size = 10,verbose = 0,validation_split = 0.1)
```

```python
#학습과정 시각화
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(model_1d_hist.history["loss"],label="train_loss")
plt.plot(model_1d_hist.history["val_loss"],label="val_loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()

plt.subplot(1,2,2)
plt.plot(model_1d_hist.history["loss"][10:],label="train_loss")
plt.plot(model_1d_hist.history["val_loss"][10:],label="val_loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend();
```
